```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt


vigs = []

for i in range(6):
    df = pd.read_csv(f"C:\\Users\\luktu\\Downloads\\gawll (1)\\class_{i}\\eVIG_dataset_urban_class_{i}_c1_a1_r1.csv", header=None).values
    vigs.append(df)

agg_vig = np.asarray(vigs).sum(axis=0)

G = nx.Graph()
for i in range(agg_vig.shape[0]):
    for j in range(i + 1, agg_vig.shape[1]):
        if agg_vig[i][j] > agg_vig[i].mean() + 2 * agg_vig[i].std():
            G.add_edge(i, j, weight=agg_vig[i][j])


fig, ax = plt.subplots(figsize=(20, 12))  # Adjust the width and height as needed
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10, font_color='black', font_weight='bold', edge_color='gray')
plt.show()

# pass

```
