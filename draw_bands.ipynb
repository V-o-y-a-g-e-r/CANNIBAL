```python
import numpy as np
import matplotlib.pyplot as plt


data = np.load("others/pavia.npy")

for i in range(30):
    bands = np.loadtxt(
        f"random_pavia_results_test_frac_band_sel_estimators_4/bands---n_clusters-5,test_size-band_sel,random_state-0.txt"
    ).astype(int)
    bands = np.sort(np.random.choice(bands, size=3, replace=False))
    data_sub = data[..., bands]
    data_sub = (data_sub - data_sub.min()) / (data_sub.max() - data_sub.min())

    # Create a larger figure with increased DPI
    fig, ax = plt.subplots(figsize=(10, 10), dpi=1000)

    # Display the normalized image
    ax.imshow(data_sub)

    # Remove white margins
    ax.set_axis_off()

    # Save the plot without margins
    plt.savefig(f"random_{bands.tolist()}_PAVIA.pdf", bbox_inches="tight", pad_inches=0)
    plt.show()
    pass
```
